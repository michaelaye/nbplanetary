```python
#| export
from yarl import URL
from planetarypy.pds.indexes import Index
```

```python
url = URL("https://pds-geosciences.wustl.edu/mro/mro-m-crism-5-rdr-mptargeted-v1/mrocr_4001/index/mtrdr0705_index.lbl")
```

```python
#| export
class MTRDRIndex(Index):
    def __init__(self, url):
        super().__init__(key='mro.crism.mtrdr', url=url)
```

```python
index = MTRDRIndex(url=url)
```

```python
index.download()
```

```python
index.convert_to_parquet()
```
