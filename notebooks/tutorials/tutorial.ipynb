# Tutorial

This is a general tutorial for many of the features `planetarypy` offers.

## Main features for data access
The main idea for access to archived planetary data is that retrieval of instrument data should be as simple as naming a mission, an instrument, potentially a data level, and a data ID.
To this end, a dot-separated code can be used to access data catalogs from the Planetary Data System (PDS).
If you don't know what instruments are supported in `planetarypy` (a list that could grow with your help), you can use the `find_instruments` function:

```python
from planetarypy.pds.apps import find_indexes, get_index, find_instruments
```

```python
find_instruments?
```

```python
find_instruments("mro")
```

Now, seeing that the CTX instrument is supported, one could ask what index files (= catalogs) are available, using the dotted code for `mission.instrument`:

```python
find_indexes("mro.ctx")
```

We only have one index file for EDR data. Let's get it, using the now 3-segments dot-code (it returns a `pandas` Dataframe):

```python
index = get_index('mro.ctx.edr', refresh=False)
index.head()
```

You might have ssen, that `planetarypy` realized that the existing index on my disk was outdated and proceeded to get an updated one.

The warning is about badly formatted data on the PDS for which `planetarypy` had to force path names to the correct letter case.
This is not something we can influence, but we reported the issue at the PDS, and here, we just let the user know, that we `fudged` around that issue.

Note that the catalog ascii times have been converted to proper datetime objects, using the utility functions in `planetarypy.utils`.
This means that you can use these times to create proper plots over time in case you are interested in catalog statistics.

```python
index.IMAGE_TIME
```

Using standard [pandas operations](https://pandas.pydata.org/docs/), we can now filte the data down, and identify a data product we are interested in:

```python
index.columns
```

```python
index.EMISSION_ANGLE.describe()
```

```python
index.LOCAL_TIME.describe()
```

```python
angle_filter = index.EMISSION_ANGLE < 1.0
```

Let's also calculate a mean latitude and longitude instead of the image corners, for rough location filtering.
As a habit, we recommend to assign lower case letters to column names that we add on top of the PDS ones:

```python
index = index.assign(
    lat_mean=(
        index.UPPER_LEFT_LATITUDE
        + index.UPPER_RIGHT_LATITUDE
        + index.LOWER_LEFT_LATITUDE
        + index.LOWER_RIGHT_LATITUDE) /
    4)
```

```python
index = index.assign(
    lon_mean=(
        index.UPPER_LEFT_LONGITUDE
        + index.UPPER_RIGHT_LONGITUDE
        + index.LOWER_LEFT_LONGITUDE
        + index.LOWER_RIGHT_LONGITUDE) /
    4)
```

```python
index.info()
```

```python
lat_filter = index.lat_mean.between(-80,-60)
lon_filter = index.lon_mean.between(0, 10)
```

```python
index[angle_filter & lat_filter & lon_filter]
```

Copying a `PRODUCT_ID`, we can now automatically download it via the CTX utilities:

```python
pid = "P03_002074_1155_XN_64S004W"
```

```python
from planetarypy import ctx
```

Above could issue a warning that is about not finding an ISIS installation, which is required for _SOME_ of this library's features.
Make sure your current environment has the environment variable ISISROOT and ISISDATA defined, so that planetarypy can find the ISIS tools, in case you want to use functions related to them.

Now let's use the CTX class to access data:

```python
data = ctx.CTX(pid)
```

```python
data.download()
```

The data was downloaded to a mission- and instrument-based storage folder that was defined the first time you imported `planetarypy`.
You can verify it's location by looking at the config object:

```python
from planetarypy.config import config
```

```python
config.storage_root
```

We can now read the data into a multi-dimensional xarray, which supports elaborate indexing via its coordinates.

```python
data.edr_da
```

But one also get the image data simply as a numpy array:

```python
data.edr_da.values
```

Or plot the image using holoviews and datashader, so that even large images won't bog down your browser:

```python
data.plot_edr()
```

Note, that the above data is the uncalibrated EDR, so the dynamics of the image seem a little rough.
